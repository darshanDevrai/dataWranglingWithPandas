## Analyzation of #WeRateDogs twitter data

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.plotting import scatter_matrix
```

### Lets take a look at our cleaned #WeRateDogs twitter data

```python
df = pd.read_csv("twitter_archive_master.csv", index_col=0)
df.info()
```

```python
### Lets see the Stats about the numerical variables in the dataset.
df.describe()
```

Here, 
 - the mean of the rating column is 11.57 with max outlier 1176. 
 - the mean of the retweet_count column is 2766.75 with max count 79515 and min count 16.
 - the mean of the favorite_count column is 8895.72 with max count 132810 and min count 81.
 - the breed_confidence(highest) generated by the Neural network has mean of 0.464991 with min 0 and max 0.99.

### Lets see the retweet_count and favorite_count over time

```python
# It seems that when we store the dataframe to csv the datetime object is not saved at it is. Instead it is 
# saved as an object from this url https://stackoverflow.com/questions/30298144/datetime-format-change-when-save-to-csv-file-python.
# So the cleaning step of converting timestamp to datetime is useless, still I am going to keep it.
# Convert timestamp column from object to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

df.set_index('timestamp', inplace=True)
```

```python



df['retweet_count'].plot(color = 'tab:orange', label='Retweets')
df['favorite_count'].plot(color = 'tab:purple', label='Favorites')
plt.style.use('seaborn-talk')
plt.legend(loc='upper right')
plt.xlabel('Timestamp')
plt.ylabel('Count')
plt.title('Retweets_counts and favorites_count over time')
# plt.savefig('retweets_favorites.png')
plt.show()
```

Here, we can see that retweets and favorites counts are increasing over time

***

### Lets see the corelation between different variables in the data

#### Lets use pandas corr() function and use scatter_matrix which is similar to ggpairs in R which we have learned in EDA lesson.

```python
df.corr(method='pearson')
```

```python
df[['rating','breed_confidence','retweet_count','favorite_count']].corr(method='pearson')
```

```python
scatter_matrix(df[['rating','breed_confidence','retweet_count','favorite_count']], alpha=0.1, diagonal='kde')
```

```python
df.plot(kind='scatter', x='retweet_count', y='favorite_count',color = 'tab:pink')
plt.xlabel('retweet_count')
plt.ylabel('favorite_count')
plt.title('Retweets_counts vs favorites_count')
plt.show()
```

From the above statistics and plot we can see that there is a strong positive corelation between retweet counts and favorite counts. It means that people make such tweets favorite which gets retweeted more. It makes sense because people make retweets which they find interesting and the tweet gets circulated more resulting such tweets get more favorite count.

### Which breed of dog gets higher ratings.

```python
breed_df = df.copy()
# Lets remove outliers i.e make the criteria for higher ratings between 11 to 17
breed_df[(breed_df.rating > 11.0) & (breed_df.rating < 17.0) & (breed_df.breed != 'unknown')].breed.value_counts()
```

From this stats above, golden_retriever gets highest ratings very frequently this is maybe bacause #WeRateDogs gets golden_retriever's pic more often. But this is not the case for Labrador_retriever and pug because we know from before that Labrador_retriever's pic was send most frequent and pug's pic was fifth most frequent but, Pembroke gets higher ratings than Labrador_retriever though the ocurance of Pembroke is less than Labrador_retriever. Similar is the case with Pug and some other dogs.
